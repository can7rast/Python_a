```python
def len_of_key(D):
    new_dict = {}
    lens = []

    #вычисляем кол-во различных длин ключей
    for i in D.keys():
        lens.append(len(i))
    lens = set(lens)

    #заполняем новый словарь кол-вом lists
    for i in lens:
        new_dict.setdefault(i, [])

    #заполняем ключи значениями
    for i in D.keys():
        new_dict[len(i)].append(i)

    return new_dict

d = {"1234" : 1234, "5678" : 1234, "231" : 123, "12" : 1}
print(len_of_key(d))
```

```python
from functools import reduce

def inter_set(list_set):
    return reduce(lambda x, y: x.intersection(y), list_set)

set1 = {1, 2, 3, 4}
set2 = {2, 3, 4, 5}
set3 = {3, 4, 5, 6}
sets = [set1, set2, set3]

print(inter_set(sets))
```
