Задание 3

Даны три положительных числа a, b, c. Проверить будут ли они
сторонами треугольника? Если да, то вычислить площадь этого треугольника.

```python
def third_task(a,b,c):
    result = 0
    if a+b > c and a + c > b  and b + c > a:
        p = (a+b+c) / 2
        result = (p*(p-a)*(p-b)*(p-c)) ** 0.5
    return result


a, b, c = map(float,input("Введите 3 стороны:").split())
result = third_task(a,b,c)
print(result)
```

Задание 4

Известны сопротивления двух несоединенных друг с другом участков
электрической цепи и напряжение на каждом из них. По какому участку
протекает меньший ток?

```python
def fourth_task(r1,r2,u1,u2):
    if u1/r1 < u2/r2:
      return 1
    elif u1/r1 > u2/r2:
      return 2
    else:
      print("Ток равен")
      return 0


r1, r2 = map(float, input("Введите сопротивление двух участков разных цепей:").split())
u1, u2 = map(float, input("Введите напряжение на этих участках цепей:").split())

print(f'Наименьший ток на {fourth_task(r1,r2,u1,u2)} участке')
```

Задание 5

Дата некоторого дня характеризуется двумя натуральными числами: m
(порядковый номер месяца) и n (число). По заданным n и m определить:
а) дату предыдущего дня (принять, что n и m не характеризуют 1 января);
б) дату следующего дня (принять, что n и m не характеризуют 31 декабря)

```python
def fifth_task(m,n):
    array1 = [1,3,5,7,8,10,12]
    array2 = [2,4,6,9,11]

    past_day = 0
    next_day = 0

    feb_past = 0 if m != 3 else 1
    if m in array1:
        if n != 1:
            past_day = n - 1
        elif n == 1 and not feb_past:
            past_day = 30
        else:
            past_day = 28

        if n == 31:
            next_day = 1
        else:
            next_day = n + 1


    if m in array2:
        if n != 1:
            past_day = n - 1
        elif n == 1:
            past_day = 31

        if n == 30:
            next_day = 1
        elif n == 28 and m == 2:
            next_day = 1
        else:
            next_day = n + 1
    print("Следующий день:", next_day)
    print("Вчерашний день:",past_day)


a, b = map(int,input("Введите месяц и день:").split())
fifth_task(a,b)
```

```python
import math

eps = 1e-3
sum = 0
stop = 0
n = 1
while stop == 0:
    a_N = math.factorial(n) / (n ** n)
    sum += a_N
    if (abs(a_N) < eps):
        stop = 1
    n += 1
print(sum)
```

```python
def max_dlina(ar):
    d_x = 0
    d_y = 0
    res = 0
    for i in range(len(ar)):
        for j in range(i+1,len(ar)):
            d_x = abs(ar[i][0] - ar[j][0])
            d_y = abs(ar[i][1] - ar[j][1])
            temp = (pow(d_x, 2) + pow(d_y, 2)) ** 0.5
            if temp > res:
                res = temp
    return res


ar = []

for i in range(int(input("Введите кол-во точек:"))):
    ar.append(list(map(int,input("Введите x и y для точки:").split())))

print(max_dlina(ar))
```
