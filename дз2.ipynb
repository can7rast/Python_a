```python
class invoice:
    def __init__(self, number, sum):
        self.__number = number
        self.__sum = sum

    @property
    def number(self):
        return self.__number

    @property
    def sum(self):
        return self.__sum

    @number.setter
    def number(self, value):
        self.__number = value

    @sum.setter
    def sum(self, value):
        if value < 0:
            print("Sum is negative")
        else:
            self.__sum = value

    def update_amount(self, amount):
        if amount < 0:
            print("Sum is negative")
        else:
            self.__sum += amount

    def print_info(self):
        print(f"Number: {self.number}")
        print(f"Sum: {self.sum}")



a = invoice(808088, 1000)

a.print_info()

a.update_amount(100)

print("Новые данные:")
a.print_info()
```

```python
import abc
class wheel(abc.ABC):
    @abc.abstractmethod
    def diameter(self):
        pass

class bike(wheel):
    def __init__(self):
        self.__diameter = None
        self.__wheels = []

    @property
    def diameter(self):
        return self.__diameter

    @diameter.setter
    def diameter(self, diameter):
        if diameter <= 0:
            print("Not valid diameter")
        else:
            self.__diameter = diameter

    def add_wheel(self):
        self.__wheels.append(self.__diameter)

    def check(self):
        print(f"Всего {len(self.__wheels)} колес(а) с диаметром {self.__diameter}")


a = bike()
a.diameter = 1
a.add_wheel()
a.add_wheel()
a.check()
```

```python
class fitness_coach:
    def __init__(self, name, specialization):
        self.__name = name
        self.__specialization = specialization

    @property
    def name(self):
        return self.__name
    @property
    def specialization(self):
        return self.__specialization
    @name.setter
    def name(self, value):
        self.__name = value
    @specialization.setter
    def specialization(self, value):
        self.__specialization = value

    def print_info(self):
        print(f"Имя трененра:{self.__name} его специализация {self.__specialization}")


class fitnessClub:
    def __init__(self):
        self.__coaches = []

    def add_coach(self, coach):
        self.__coaches.append(coach)


    def print_coaches(self):
        for coach in self.__coaches:
            coach.print_info()


a1 = fitness_coach("Artem", "jim")
a2 = fitness_coach("Alsu", "cardio")
c = fitnessClub()
c.add_coach(a1)
c.add_coach(a2)
c.print_coaches()
```

```python
import abc
class photo(abc.ABC):
    def __init__(self, name):
        self.name = name
    def edit(self):
        pass


class Portrait(photo):
    def __init__(self, name):
        super().__init__(name)

    def edit(self, new_name):
        self.name = new_name

    def print_name(self):
        print(f"Перед вами картина с названием: {self.name}")


class LandSpace(photo):
    def __init__(self, name):
        super().__init__(name)

    def edit(self, new_name):
        self.name = new_name

    def print_name(self):
        print(f"Перед вами пейзаж с названием: {self.name}")


a = Portrait("Артем")
a.edit("Танец")
a.print_name()


b = LandSpace("Шишкин лес")
b.print_name()

```

```python
import abc

class IInvoice(abc.ABC):
    @abc.abstractmethod
    def update_amount(self, amount : float):
        pass
    @abc.abstractmethod
    def get_invoice_details(self):
        pass

class Invoice(IInvoice):
    def  __init__(self, amount, invoice):
        self.amount = amount
        self.invoiceNumber = invoice

    def update_amount(self, amount : float):
        if amount < 0:
            print("Not valid amount")
        else:
            self.amount = amount

    def get_invoice_details(self):
        print(f"Номер счета {self.invoiceNumber} с балансом {self.amount}")


a = Invoice(1000, 123)
a.update_amount(100)
a.get_invoice_details()
```
