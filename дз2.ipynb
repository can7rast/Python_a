```python
class invoice:
    def __init__(self, number, sum):
        self.__number = number
        self.__sum = sum

    @property
    def number(self):
        return self.__number

    @property
    def sum(self):
        return self.__sum

    @number.setter
    def number(self, value):
        self.__number = value

    @sum.setter
    def sum(self, value):
        if value < 0:
            print("Sum is negative")
        else:
            self.__sum = value

    def update_amount(self, amount):
        if amount < 0:
            print("Sum is negative")
        else:
            self.__sum += amount

    def print_info(self):
        print(f"Number: {self.number}")
        print(f"Sum: {self.sum}")



a = invoice(808088, 1000)

a.print_info()

a.update_amount(100)

print("Новые данные:")
a.print_info()
```
