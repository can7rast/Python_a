```python
def str_to_sum(s):
    sum = 0
    for i in s:
        sum += ord(i)
    return sum

print(str_to_sum("SUM"))
```

```python
def read_string():
    k = ""
    sign = input("Введите знак:")
    if sign not in ["+", "-"]:
        print("Введите + или -!")
        return

    c = ""
    while c != " ":
        c = input("Введите символ:")

        if len(c) != 1:
            print("Нужно вводить по одному символу!")
            return

        k += c

    k = int(k)
    if sign == "+":
        k *= 1
    else:
        k *= -1
    return k

r = read_string()
print(r)
```

```python
from pathlib import Path as p

def file_sort(path):

    with open(path) as f:
        arr = []
        for line in f:
            arr.append(line.rstrip("\n"))
        sorted_arr = sorted(arr, key= lambda x: len(x),reverse=True)
        new_file = p(input("Enter new file name: ")).with_suffix(".txt")
        with new_file.open("w") as writer:
            for line in sorted_arr:
                writer.write(line + '\n')

path = input("Введите название файла:")
file_sort(path)
```

```python
def clear_file(path):
    with open(path, 'r+') as f:
        arr = f.readlines()
        n = len(arr) // 2
        if len(arr) > 0:
            f.seek(0)
            f.writelines(arr[:n])
            f.truncate()
            print("Вторая половина файла удалена")
        else:
            print("Кол-во строк не подходит")
    return




p = input("Введите название файла:")
clear_file(p)
```

```python
def len_line(path):
    with open(path) as f:
        line = f.readline().strip("\n\t").split()
        n = 1
        for i in range(len(line) - 1):
            if line[i] != line[i+1]:
                n += 1
        return n


p = input("Введите название файла:")
print(len_line(p))
```
